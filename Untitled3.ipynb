```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Step 1: Data Synthesis
num_samples = 1000  # Increase the number of samples
image_shape = (224, 224, 3)
text_max_length = 50  

# Synthetic image data
images = np.random.rand(num_samples, *image_shape)

# Synthetic text data
text_data = ['Beautiful gold necklace with diamonds',
             'Elegant silver earrings with sapphires',
             'Stylish rose gold bracelet with rubies',
             'Exquisite diamond engagement ring',
             'Chic pearl pendant necklace',
             'Vintage emerald stud earrings',
             'Modern titanium cufflinks',
             'Classic ruby and diamond ring',
             'Trendy leather wrap bracelet',
             'Funky geometric earrings']

# Synthetic labels
labels = np.random.choice(['necklace', 'earrings', 'bracelet', 'ring', 'pendant', 'cufflinks'], num_samples)
labels_int = np.array([{'necklace': 0, 'earrings': 1, 'bracelet': 2, 'ring': 3, 'pendant': 4, 'cufflinks': 5}[label] for label in labels])

# Step 2: Model Selection
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=image_shape),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),  # Add dropout
    tf.keras.layers.Dense(6, activation='softmax')  # Change num_classes to 6
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Step 3: Model Training
history = model.fit(images, labels_int, epochs=10, batch_size=32, validation_split=0.2)  # Increase batch size

# Plot training history
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Step 4: Model Testing
# Generate synthetic test data
test_images = np.random.rand(num_samples, *image_shape)
test_labels = np.random.choice(['necklace', 'earrings', 'bracelet', 'ring', 'pendant', 'cufflinks'], num_samples)
test_labels_int = np.array([{'necklace': 0, 'earrings': 1, 'bracelet': 2, 'ring': 3, 'pendant': 4, 'cufflinks': 5}[label] for label in test_labels])

# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(test_images, test_labels_int, verbose=2)

print(f'Test Accuracy: {test_accuracy}')
```
